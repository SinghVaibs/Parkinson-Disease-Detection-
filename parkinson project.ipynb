```python
import cv2
import numpy as np
import pandas as pd
from skimage.io import imread
import keras.layers as layers
import tensorflow as tf
import random
import keras
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator

IMG_HEIGHT = 256
IMG_WIDTH = 256
IMG_CHANNELS = 1


X_train = np.zeros((69, IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS), dtype=np.uint8)
Y_train = np.zeros(69, dtype=np.uint8)

for i in range(1, 37):
    im = imread('/Users/vaibhav/Downloads/IMAGES/P({}).png'.format(i))
    X_train[i-1, :, :, 0] = im[:, :, 0]
    Y_train[i-1] = 1

for i in range(1, 34):
    im = imread('/Users/vaibhav/Downloads/IMAGES/H({}).png'.format(i))
    X_train[i+36-1, :, :, 0] = im[:, :, 0]


plt.imshow(X_train[54, :, :, 0], cmap='gray')
plt.show()


shuffler = np.random.permutation(len(X_train))
X_train = X_train[shuffler]
Y_train = Y_train[shuffler]
X_train = X_train.astype("float32") / 255

num_classes = 2
input_shape = (256, 256, 1)


model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(2, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(4, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(6, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(8, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(10, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(12, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dropout(0.4),
        layers.Dense(12, activation="relu"),
        layers.Dropout(0.4),
        layers.Dense(1, activation="sigmoid"),        
    ]
)

model.summary()


data_generator = ImageDataGenerator(
    rotation_range=25,
    horizontal_flip=True,
    width_shift_range=0.02,
    height_shift_range=0.02,
)
data_generator.fit(X_train)


model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
history = model.fit_generator(data_generator.flow(X_train, Y_train), epochs=200)


plt.plot(history.history['loss'])


X_test = np.zeros((30, IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS), dtype=np.uint8)
Y_test = np.zeros(30, dtype=np.uint8)

for i in range(1, 16):
    im = imread('/Users/vaibhav/Downloads/IMAGES/P({}).png'.format(i))
    X_test[i-1, :, :, 0] = im[:, :, 0]
    Y_test[i-1] = 1

for i in range(1, 16):
    im = imread('/Users/vaibhav/Downloads/IMAGES/H({}).png'.format(i))
    X_test[i+15-1, :, :, 0] = im[:, :, 0]


plt.imshow(X_test[14, :, :, 0], cmap='gray')
plt.show()


shuffler = np.random.permutation(len(X_test))
X_test = X_test[shuffler]
Y_test = Y_test[shuffler]
X_test = X_test.astype("float32") / 255


data_generator = ImageDataGenerator(
    rotation_range=20,
    horizontal_flip=True,
    width_shift_range=0.02,
    height_shift_range=0.02,
)
data_generator.fit(X_test)


score = model.evaluate(data_generator.flow(X_test, Y_test), verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])


img = X_test[2]
img = np.expand_dims(img, axis=0)
fig = plt.figure(figsize=(20, 40))

plt_count = 1
filt_count = 2
for i in range(1, 12, 2):
    model_feature = keras.Model(inputs=model.inputs, outputs=model.layers[i].output)
    feature_maps = model_feature.predict(img)
    for j in range(filt_count):
        fig.add_subplot(11, 4, plt_count)
        plt.imshow(feature_maps[0, :, :, j])
        plt_count += 1
    filt_count += 2

plt.show()

y_pred = model.predict(X_test)
res = [1 if y >= 0.5 else 0 for y in y_pred]


plt.imshow(X_test[14, :, :, 0], cmap='gray')
if res[14] == 0:
    print("Healthy Person")
else:
    print("Parkinson Detected")


from sklearn.metrics import precision_score, recall_score, confusion_matrix, roc_curve, auc

precision = precision_score(Y_test, res)
recall = recall_score(Y_test, res)
cf_matrix = confusion_matrix(Y_test, res)

print('Precision:', precision)
print('Recall:', recall)
print('Confusion Matrix:')
print(cf_matrix)


fpr, tpr, thresholds = roc_curve(Y_test, y_pred)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve')
plt.legend(loc='lower right')
plt.show()
```
